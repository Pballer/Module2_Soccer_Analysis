We checked the VIEW command used to create the FlatView_Advanced joined table. But we would like to keep our data cleaning in python rather than sqlite so we created a large dataframe that has not been filtered down to the the 2011. This is to make sure that there are no errors between the 2011 label and the actual dates (5th of August, 2011 to the 5th of May, 2012) of the season. (https://en.wikipedia.org/wiki/2011%E2%80%9312_Bundesliga)

```python
cursor.execute('''SELECT Match_ID,
                Unique_Team_ID, 
                Unique_Team, 
                Div, Season, 
                Date, 
                HomeTeam, 
                AwayTeam, 
                FTHG, 
                FTAG, 
                FTR 
                FROM FlatView_Advanced''')
all_variables_raw_unfiltered = pd.DataFrame(cursor.fetchall())
all_variables_raw_unfiltered.columns = [x[0] for x in cursor.description]
all_variables_raw_unfiltered.head()
```

```python
all_variables_raw_unfiltered.Date = pd.to_datetime(all_variables_raw_unfiltered.Date)
```

```python
all_variables_raw_unfiltered[all_variables_raw_unfiltered['Season'] == '2011'].Date.describe().T
```

```python
all_variables_filtered = all_variables_raw_unfiltered.loc[((all_variables_raw_unfiltered.Div == 'D1')
                                 |(all_variables_raw_unfiltered.Div == 'D2')) 
                                 & (all_variables_raw_unfiltered.Season == '2011')]
all_variables_filtered
```
